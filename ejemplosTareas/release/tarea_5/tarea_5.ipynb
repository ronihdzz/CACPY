```python
###################################################################################################################
# NO MODIFICAR EL CODIGO DE LA CLASE: Trabajador
##################################################################################################################

class Trabajador():
    def __init__(self,nombre,edad,sueldo):
        '''
        Este el metodo constructor de las clase: Trabajador, el 
        cual se encargara de construir el objeto con las caracteristicas
        especificadas en los parametros.
            
        Parametros:
        nombre (str): Nombre del trabajador
        edad (int): Edad en aos del trabajador
        sueldo (float): Sueldo del trabajador

        Returns:
        Una instancia de la clase: 'Trabajador'
        '''
        
        self.nombre=nombre
        self.edad=edad
        self.sueldo=sueldo
        self.numeroJornadas_sinCobrar=0
        self.ahorro=0
    
    def trabajar(self):
        '''
        Este metodo aumentara en uno, el numero numero de jornadas de
        trabajo que el trabajador a hecho pero aun no cobra, por tal
        motivo dicho metodo aumentara en uno el valor del atributo
        de instancia 'self.numeroJornadas_sinCobrar'.
        
        Este metodo al llamarse tambien imprimira un mensaje de felicitacin
        al trabajador por haber laborado
        '''
        
        self.numeroJornadas_sinCobrar+=1
        print("Excelente trabajad@r: ",self.nombre," estamos orgullosos de que trabajes")
        
    def cobrar(self):
        '''
        Este metodo hara que el trabajador cobre el numero de jornadas
        de trabajo que a hecho hasta el momento pero que aun no cobra
        '''
        
        if self.numeroJornadas_sinCobrar>0:
            # Calculando el monto de dinero que se le debe pagar al trabajador
            dineroDebeAlTrabajador=self.numeroJornadas_sinCobrar*self.sueldo
            
            # Pagandole el monto al trabajador
            self.ahorro=self.ahorro+dineroDebeAlTrabajador
            
            self.numeroJornadas_sinCobrar=0
            
            print("Trabajad@r: ",self.nombre," su pago a sido realizado")
            
        else:
            print("Trabajad@r: ",self.nombre," no se le debe ninguna jornada de trabajo")
            
    def comprar(self,precioDelProducto):
        '''
        Este metodo permitira al trabajador comprar el producto
        que desea comprar, siempre y cuando le alcance
        '''
        
        if self.ahorro>=precioDelProducto:
            self.ahorro=self.ahorro-precioDelProducto
            
            print("El producto ha sido comprado exitosamente :)")
        
        else:
            print("No se tiene el dinero suficiente para comprar el producto :)")
            
```
